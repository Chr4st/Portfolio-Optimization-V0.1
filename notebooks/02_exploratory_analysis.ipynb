```python
# 02_exploratory_analysis.ipynb

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# ✅ Load cleaned portfolio data
data_path = "../data/raw/my_actual_portfolio.csv"
data = pd.read_csv(data_path, index_col=0, parse_dates=True)

# ✅ Plot the normalized price trend (all start at 1)
normalized = data / data.iloc[0]
normalized.plot(figsize=(14, 6), title="Normalized Portfolio Prices")
plt.ylabel("Normalized Price")
plt.grid(True)
plt.show()

# ✅ Calculate daily returns
returns = data.pct_change().dropna()

# ✅ Plot histogram of returns
returns.hist(figsize=(14, 8), bins=50)
plt.suptitle("Distribution of Daily Returns", fontsize=16)
plt.show()

# ✅ Correlation heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(returns.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix of Asset Returns")
plt.show()

# ✅ Save returns for use in optimization
os.makedirs("../data/processed/", exist_ok=True)
returns.to_csv("../data/processed/portfolio_returns.csv")
print("✅ Daily returns saved to '../data/processed/portfolio_returns.csv'")
```
